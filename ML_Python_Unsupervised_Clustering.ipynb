```python
import pandas as pd

data=pd.DataFrame({'x':[12,24,28,33,18,29,52,45,24,55,51,61,53,69,72,64,49,58],
                  'y':[36,39,30,52,54,46,55,59,63,70,66,63,58,23,14,8,19,7]})

data
```

```python
import matplotlib.pyplot  as plt
plt.scatter(data['x'] , data['y'])
plt.show()
```

```python
print(data['x'].min(),data['x'].max())
print(data['y'].min(),data['y'].max())
```

```python
import numpy as np
cent0 = [np.random.randint(12,72) , np.random.randint(7,70)]
cent1 = [np.random.randint(12,72) , np.random.randint(7,70)]
cent2 = [np.random.randint(12,72) , np.random.randint(7,70)]

print(cent0,cent1,cent2)
```

```python
import matplotlib.pyplot  as plt

plt.scatter(data['x'] , data['y'])

plt.scatter(cent0[0] , cent0[1] , marker='*' , color='r')
plt.scatter(cent1[0] , cent1[1] , marker='*' , color='g')
plt.scatter(cent2[0] , cent2[1] , marker='*' , color='b')

plt.show()
```

```python
data['r'] = ((data['x'] - cent0[0])**2 + (data['y'] - cent0[1])**2)**0.5
data['g'] = ((data['x'] - cent1[0])**2 + (data['y'] - cent1[1])**2)**0.5
data['b'] = ((data['x'] - cent2[0])**2 + (data['y'] - cent2[1])**2)**0.5
```

```python
data.head()
```

```python
data['minimum'] = data[['r','g','b']].idxmin(axis=1)
```

```python
data.head()
```

```python
import matplotlib.pyplot  as plt

plt.scatter(data['x'] , data['y'] , c = data['minimum'])

plt.scatter(cent0[0] , cent0[1] , marker='*' , color='r')
plt.scatter(cent1[0] , cent1[1] , marker='*' , color='g')
plt.scatter(cent2[0] , cent2[1] , marker='*' , color='b')

plt.show()
```

```python
cent0[0] = data[data['minimum'] == 'r']['x'].mean()
cent0[1] = data[data['minimum'] == 'r']['y'].mean()

cent1[0] = data[data['minimum'] == 'g']['x'].mean()
cent1[1] = data[data['minimum'] == 'g']['y'].mean()

cent2[0] = data[data['minimum'] == 'b']['x'].mean()
cent2[1] = data[data['minimum'] == 'b']['y'].mean()
```

```python
import matplotlib.pyplot  as plt

plt.scatter(data['x'] , data['y'] , c = data['minimum'])

plt.scatter(cent0[0] , cent0[1] , marker='*' , color='r')
plt.scatter(cent1[0] , cent1[1] , marker='*' , color='g')
plt.scatter(cent2[0] , cent2[1] , marker='*' , color='b')

plt.show()
```

```python
data['r'] = ((data['x'] - cent0[0])**2 + (data['y'] - cent0[1])**2)**0.5
data['g'] = ((data['x'] - cent1[0])**2 + (data['y'] - cent1[1])**2)**0.5
data['b'] = ((data['x'] - cent2[0])**2 + (data['y'] - cent2[1])**2)**0.5

data['minimum'] = data[['r','g','b']].idxmin(axis=1)
```

```python
import matplotlib.pyplot  as plt

plt.scatter(data['x'] , data['y'] , c = data['minimum'])

plt.scatter(cent0[0] , cent0[1] , marker='*' , color='r')
plt.scatter(cent1[0] , cent1[1] , marker='*' , color='g')
plt.scatter(cent2[0] , cent2[1] , marker='*' , color='b')

plt.show()
```

# customer classification

```python
df = pd.read_csv('mall.csv')
df.head()
```

```python
df.columns
```

```python
df = df[['Age','Annual Income (k$)','Spending Score (1-100)']]
df.head()
```

```python
from sklearn.cluster import KMeans
losses = []
for i in range(1,11):
  kmodel = KMeans(n_clusters=i)
  kmodel.fit(df)

  losses.append(kmodel.inertia_)
```

```python
plt.plot(range(1,11) , losses)
plt.show()
```

```python
from sklearn.cluster import KMeans
kmodel = KMeans(5)
kmodel.fit(df)

df['Label'] = kmodel.predict(df)
```

```python
df.head()
```

```python
kmodel.cluster_centers_
```

```python
plt.scatter(df['Annual Income (k$)'] , df['Spending Score (1-100)'] , c=df['Label'])
plt.show()
```
